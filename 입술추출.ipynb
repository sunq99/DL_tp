```python
import os
import cv2
import torch
import ffmpeg
from pipelines.data.data_module import AVSRDataLoader
from pipelines.detectors.mediapipe.detector import LandmarksDetector

def save2vid(filename, vid, frames_per_second):
    os.makedirs(os.path.dirname(filename), exist_ok=True)

    if not isinstance(vid, torch.Tensor):
        raise TypeError("vid must be a torch.Tensor")
    if vid.dtype != torch.uint8:
        vid = vid.to(torch.uint8)  # Convert to uint8
    if vid.dim() != 4 or vid.shape[-1] != 3:
        raise ValueError("vid must have shape (T, H, W, C) with 3 channels")

    T, H, W, C = vid.shape

    process = (
        ffmpeg
        .input('pipe:', format='rawvideo', pix_fmt='rgb24', s=f'{W}x{H}', r=frames_per_second)
        .output(filename, pix_fmt='yuv420p', vcodec='libx264', r=frames_per_second)
        .overwrite_output()
        .run_async(pipe_stdin=True)
    )

    for frame in vid.numpy():
        process.stdin.write(frame.tobytes())

    process.stdin.close()
    process.wait()

def preprocess_video(src_filename, dst_filename):
    landmarks = landmarks_detector(src_filename)
    data = dataloader.load_data(src_filename, landmarks)

    fps = cv2.VideoCapture(src_filename).get(cv2.CAP_PROP_FPS)
    fps_int = int(round(fps))  # Ensure fps is an integer

    save2vid(dst_filename, data, fps_int)

def preprocess_video_folder(src_folder, dst_folder, file_extensions=('.mp4', '.avi', '.mov', '.mkv')):
    """í´ë” ë‚´ ëª¨ë“  ë™ì˜ìƒì„ ì²˜ë¦¬í•˜ì—¬ ì €ì¥"""
    os.makedirs(dst_folder, exist_ok=True)

    for filename in os.listdir(src_folder):
        if filename.lower().endswith(file_extensions):
            src_path = os.path.join(src_folder, filename)
            dst_path = os.path.join(dst_folder, filename)

            print(f"Processing: {src_path} -> {dst_path}")
            preprocess_video(src_path, dst_path)

# ë°ì´í„° ë¡œë” ë° ëœë“œë§ˆí¬ ê°ì§€ê¸° ì´ˆê¸°í™”
dataloader = AVSRDataLoader(modality="video", speed_rate=1, transform=False, detector="mediapipe", convert_gray=False)
landmarks_detector = LandmarksDetector()

# # # í´ë” ë‚´ ëª¨ë“  ë¹„ë””ì˜¤ ì²˜ë¦¬ ì‹¤í–‰ ì˜ˆì‹œ
# src_folder = "/content/doit"  # ì›ë³¸ ë™ì˜ìƒ í´ë”
# dst_folder = "/content/doit/a"  # ì €ì¥ í´ë”
# preprocess_video_folder(src_folder, dst_folder)
```

```python
import ffmpeg
import librosa
from pydub import AudioSegment
import shutil  # os.rename ëŒ€ì‹  shutil.move ì‚¬ìš©
import os

# === [1] MP4ì—ì„œ WAV ì¶”ì¶œ (ì •í™•í•œ ê¸¸ì´ ìœ ì§€) ===
def extract_audio_exact_length(input_mp4, output_wav):
    """
    FFmpegì„ ì‚¬ìš©í•˜ì—¬ MP4ì—ì„œ ì›ë³¸ ì˜¤ë””ì˜¤ ê¸¸ì´ë¥¼ ìœ ì§€í•œ ì±„ WAVë¡œ ë³€í™˜
    """
    ffmpeg.input(input_mp4).output(
        output_wav,
        format='wav',
        acodec='pcm_s16le',
        ar=44100,
        ac=2,
        vsync=0
    ).run(overwrite_output=True)

    print(f"âœ…  WAV ì¶”ì¶œ ì™„ë£Œ: {output_wav} (44100 Hz, ìŠ¤í…Œë ˆì˜¤)")

# === [2] WAV íŒŒì¼ì„ 16kHz, ëª¨ë…¸ë¡œ ë³€í™˜ ===
def convert_wav_to_16k_mono(input_wav, output_wav):
    """
    Pydubì„ ì‚¬ìš©í•˜ì—¬ WAV íŒŒì¼ì„ 16kHz, ëª¨ë…¸ë¡œ ë³€í™˜
    """
    audio = AudioSegment.from_wav(input_wav)
    audio = audio.set_frame_rate(16000).set_channels(1)
    audio.export(output_wav, format="wav")

    print(f"âœ…  16kHz ë³€í™˜ ì™„ë£Œ: {output_wav} (16,000 Hz, ëª¨ë…¸)")

def get_audio_length(input_file):
    """
    FFmpegì„ ì‚¬ìš©í•˜ì—¬ ì˜¤ë””ì˜¤ ê¸¸ì´ë¥¼ ì •í™•í•˜ê²Œ ê°€ì ¸ì˜´
    """
    probe = ffmpeg.probe(input_file)
    duration = float(probe["format"]["duration"])
    return duration

def convert_mp4_to_wav_matching_avi_length(input_mp4, reference_avi, output_wav):
    """
    MP4 â†’ WAV ë³€í™˜ (ì˜¤ë””ì˜¤ëŠ” MP4ì—ì„œ ì¶”ì¶œ), AVI ê¸¸ì´ì— ë§ì¶° ì˜¤ë””ì˜¤ë¥¼ ì¡°ì •
    - WAV 16kHz ëª¨ë…¸ ë³€í™˜ í¬í•¨
    - AVI ê¸¸ì´ì— ë¬´ì¡°ê±´ ë§ì¶¤
    """
    temp_wav = "temp_original.wav"
    temp_wav_16k = "temp_16k_mono.wav"
    final_adjusted = "temp_final_adjusted.wav"

    # 1. MP4ì—ì„œ ì˜¤ë””ì˜¤ ì¶”ì¶œ
    extract_audio_exact_length(input_mp4, temp_wav)

    # 2. 16kHz ëª¨ë…¸ WAVë¡œ ë³€í™˜
    convert_wav_to_16k_mono(temp_wav, temp_wav_16k)

    # 3. AVI ê¸¸ì´ ê°€ì ¸ì˜¤ê¸° (í•­ìƒ AVI ê¸¸ì´ì— ë§ì¶œ ê²ƒ)
    avi_len = get_audio_length(reference_avi)
    target_length = avi_len  # AVI ê¸¸ì´ì— ë§ì¶˜ë‹¤

    current_length = get_audio_length(temp_wav_16k)

    print(f"\nğŸ“  MP4 ì˜¤ë””ì˜¤ ê¸¸ì´: {get_audio_length(input_mp4):.2f}ì´ˆ")
    print(f"ğŸ“  AVI ì˜ìƒ ê¸¸ì´ (ê¸°ì¤€) : {avi_len:.2f}ì´ˆ")
    print(f"ğŸ§  í˜„ì¬ WAV ê¸¸ì´: {current_length:.2f}ì´ˆ")

    # 4. AVI ê¸¸ì´ì— ë§ì¶° WAV ì¡°ì •
    if current_length > target_length:
        print("âœ‚ ì˜¤ë””ì˜¤ê°€ AVIë³´ë‹¤ ê¸¸ì–´ì„œ ì˜ë¼ëƒ…ë‹ˆë‹¤.")
        os.system(f"ffmpeg -y -i {temp_wav_16k} -t {target_length} {final_adjusted}")
    elif current_length < target_length:
        print("â•  ì˜¤ë””ì˜¤ê°€ AVIë³´ë‹¤ ì§§ì•„ì„œ íŒ¨ë”© ì¶”ê°€í•©ë‹ˆë‹¤.")
        pad_dur = target_length - current_length
        os.system(f"ffmpeg -y -i {temp_wav_16k} -filter_complex \"[0:a]apad=pad_dur={pad_dur}:pad_len=100000[aout]\" -map \"[aout]\" -t {target_length} {final_adjusted}")
    else:
        print("ğŸ‘Œ  WAV ê¸¸ì´ê°€ AVIì™€ ë™ì¼í•˜ë¯€ë¡œ ê·¸ëŒ€ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.")
        shutil.copy(temp_wav_16k, final_adjusted)

    # 5. ìµœì¢… ì €ì¥
    shutil.move(final_adjusted, output_wav)

    # 6. ìµœì¢… WAV ê¸¸ì´ í™•ì¸ ë° ì¶œë ¥
    final_length = get_audio_length(output_wav)
    print("\nğŸ“Š === ìµœì¢… ê²°ê³¼ ===")
    print(f"ğŸ“  MP4 ì˜¤ë””ì˜¤ ê¸¸ì´  : {get_audio_length(input_mp4):.2f}ì´ˆ")
    print(f"ğŸ“  AVI ì˜ìƒ ê¸¸ì´ (ê¸°ì¤€) : {avi_len:.2f}ì´ˆ")
    print(f"ğŸ§  ë³€í™˜ëœ WAV ê¸¸ì´ : {final_length:.2f}ì´ˆ")
    diff = abs(target_length - final_length)
    if diff < 0.01:
        print("ğŸ¯  ì˜¤ë””ì˜¤ ê¸¸ì´ê°€ AVI ê¸¸ì´ì™€ ì •í™•íˆ ì¼ì¹˜í•©ë‹ˆë‹¤! âœ…")
    else:
        print(f"âš ï¸  ê¸¸ì´ ì°¨ì´ ì¡´ì¬: {diff:.2f}ì´ˆ (ì¶”ê°€ í™•ì¸ í•„ìš”)")

# # === ì‹¤í–‰ ì˜ˆì‹œ ===
# input_file = "/content/drive/MyDrive/test/pink_noisy_2_BBC_cut1.mp4"  # ë³€í™˜í•  MP4 íŒŒì¼
# avi_file = "/content/drive/MyDrive/test/pink_noisy_2_BBC_cut1.avi"
# output_file = "/content/drive/MyDrive/test/pink_noisy_2_BBC_cut1.wav"
# convert_mp4_to_wav_matching_avi_length(input_file, avi_file, output_file)
```

```python
import os
import cv2
import torch
import pickle
import ffmpeg
from pipelines.data.data_module import AVSRDataLoader
from pipelines.detectors.mediapipe.detector import LandmarksDetector

def load_face_data(face_file):
    """faces.pckl íŒŒì¼ì„ ë¡œë“œí•˜ì—¬ ì–¼êµ´ ë°ì´í„° ë°˜í™˜"""
    with open(face_file, "rb") as f:
        return pickle.load(f)

def select_main_speaker(face_data, frame):
    """ì£¼ì–´ì§„ í”„ë ˆì„ì—ì„œ ê°€ì¥ ê°€ëŠ¥ì„± ë†’ì€ í™”ì ì„ íƒ"""
    faces = [entry for entry in face_data if entry['frame'] == frame]
    if not faces:
        return None

    # ê°€ì¥ ì‹ ë¢°ë„ ë†’ì€ ì–¼êµ´ ì„ íƒ
    faces.sort(key=lambda x: x['conf'], reverse=True)
    return faces[0]

def extract_lip_region(frame, landmarks):
    """Landmarksë¥¼ ì´ìš©í•´ ì…ìˆ  ì˜ì—­ì„ ì¶”ì¶œ"""
    lip_points = landmarks[48:68]  # ì…ìˆ ì˜ landmark í¬ì¸íŠ¸
    x_min, y_min = min(lip_points, key=lambda p: p[0])[0], min(lip_points, key=lambda p: p[1])[1]
    x_max, y_max = max(lip_points, key=lambda p: p[0])[0], max(lip_points, key=lambda p: p[1])[1]
    return frame[y_min:y_max, x_min:x_max]

def save2vid(filename, vid, frames_per_second):
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    if not isinstance(vid, torch.Tensor):
        raise TypeError("vid must be a torch.Tensor")
    if vid.dtype != torch.uint8:
        vid = vid.to(torch.uint8)
    if vid.dim() != 4 or vid.shape[-1] != 3:
        raise ValueError("vid must have shape (T, H, W, C) with 3 channels")

    T, H, W, C = vid.shape
    process = (
        ffmpeg
        .input('pipe:', format='rawvideo', pix_fmt='rgb24', s=f'{W}x{H}', r=frames_per_second)
        .output(filename, pix_fmt='yuv420p', vcodec='libx264', r=frames_per_second)
        .overwrite_output()
        .run_async(pipe_stdin=True)
    )

    for frame in vid.numpy():
        process.stdin.write(frame.tobytes())
    process.stdin.close()
    process.wait()

def preprocess_video(src_filename, dst_filename, face_data):
    cap = cv2.VideoCapture(src_filename)
    fps = int(round(cap.get(cv2.CAP_PROP_FPS)))
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_number = int(cap.get(cv2.CAP_PROP_POS_FRAMES)) - 1

        main_speaker = select_main_speaker(face_data, frame_number)
        if main_speaker:
            landmarks = landmarks_detector(frame)
            frame = extract_lip_region(frame, landmarks)

        frames.append(frame)

    cap.release()
    torch_frames = torch.tensor(frames)  # Tensor ë³€í™˜
    save2vid(dst_filename, torch_frames, fps)

def preprocess_video_folder(src_folder, dst_folder, face_file):
    os.makedirs(dst_folder, exist_ok=True)
    face_data = load_face_data(face_file)

    for filename in os.listdir(src_folder):
        if filename.lower().endswith(('.mp4', '.avi', '.mov', '.mkv')):
            src_path = os.path.join(src_folder, filename)
            dst_path = os.path.join(dst_folder, filename)
            print(f"Processing: {src_path} -> {dst_path}")
            preprocess_video(src_path, dst_path, face_data)

# ì‹¤í–‰ ì˜ˆì‹œ
face_file = "/content/Visual_Speech_Recognition_for_Multiple_Languages/bb.pckl"  # ì €ì¥ëœ ì–¼êµ´ ë°ì´í„° íŒŒì¼
src_folder = "input_videos"  # ì›ë³¸ ë™ì˜ìƒ í´ë”
dst_folder = "processed_videos"  # ê²°ê³¼ ì €ì¥ í´ë”
preprocess_video_folder(src_folder, dst_folder, face_file)
```

```python
import os
import cv2
import torch
import pickle
import numpy as np
import ffmpeg
from pipelines.detectors.mediapipe.detector import LandmarksDetector

# ğŸ”¹ ì–¼êµ´ ë°ì´í„° ë¡œë“œ í•¨ìˆ˜
def load_face_data(face_file):
    """faces.pckl íŒŒì¼ì„ ë¡œë“œí•˜ì—¬ ì–¼êµ´ ë°ì´í„° ë°˜í™˜"""
    try:
        with open(face_file, "rb") as f:
            data = pickle.load(f)
        print(f"ğŸ“Œ faces.pckl ë°ì´í„° ë¡œë“œ ì™„ë£Œ! ì´ {len(data)} í”„ë ˆì„ì˜ ì–¼êµ´ ì •ë³´ í¬í•¨.")
        return data
    except Exception as e:
        print(f"ğŸš¨ faces.pckl ë¡œë“œ ì˜¤ë¥˜: {e}")
        return []

# ğŸ”¹ ê°€ì¥ ì‹ ë¢°ë„ê°€ ë†’ì€ í™”ì ì„ íƒ í•¨ìˆ˜
def select_main_speaker(face_data, frame):
    """ì£¼ì–´ì§„ í”„ë ˆì„ì—ì„œ ê°€ì¥ ê°€ëŠ¥ì„± ë†’ì€ í™”ì ì„ íƒ"""
    faces = [entry for entry in face_data if isinstance(entry, dict) and 'frame' in entry and entry['frame'] == frame]
    if not faces:
        print(f"âš ï¸ í”„ë ˆì„ {frame}: ê°ì§€ëœ ì–¼êµ´ ì—†ìŒ")
        return None

    # ì‹ ë¢°ë„(conf)ê°€ ê°€ì¥ ë†’ì€ ì–¼êµ´ ì„ íƒ
    faces.sort(key=lambda x: x['conf'], reverse=True)
    print(f"âœ… í”„ë ˆì„ {frame}: ì„ íƒëœ í™”ì bbox {faces[0]['bbox']} (ì‹ ë¢°ë„: {faces[0]['conf']:.3f})")
    return faces[0]

# ğŸ”¹ ì…ìˆ  ì˜ì—­ ì¶”ì¶œ í•¨ìˆ˜
def extract_lip_region(frame, landmarks):
    """Landmarksë¥¼ ì´ìš©í•´ ì…ìˆ  ì˜ì—­ì„ ì¶”ì¶œ"""
    if landmarks is None or len(landmarks) < 68:
        print("âš ï¸ ëœë“œë§ˆí¬ ë°ì´í„° ë¶€ì¡±, ì…ìˆ  ì¶”ì¶œ ë¶ˆê°€")
        return None

    lip_points = landmarks[48:68]  # ì…ìˆ ì˜ landmark í¬ì¸íŠ¸ (Mediapipe ê¸°ì¤€)

    # ì…ìˆ  ì¢Œí‘œ ì¶”ì¶œ
    x_min, y_min = min(lip_points, key=lambda p: p[0])[0], min(lip_points, key=lambda p: p[1])[1]
    x_max, y_max = max(lip_points, key=lambda p: p[0])[0], max(lip_points, key=lambda p: p[1])[1]

    if x_max <= x_min or y_max <= y_min:
        print("âš ï¸ ì…ìˆ  ì¢Œí‘œ ì´ìƒ, í¬ë¡­ ë¶ˆê°€ëŠ¥")
        return None

    return frame[y_min:y_max, x_min:x_max]

# ğŸ”¹ ë¹„ë””ì˜¤ ì €ì¥ í•¨ìˆ˜
def save2vid(filename, vid, frames_per_second):
    os.makedirs(os.path.dirname(filename), exist_ok=True)
    if not isinstance(vid, torch.Tensor):
        raise TypeError("vid must be a torch.Tensor")
    if vid.dtype != torch.uint8:
        vid = vid.to(torch.uint8)
    if vid.dim() != 4 or vid.shape[-1] != 3:
        raise ValueError("vid must have shape (T, H, W, C) with 3 channels")

    T, H, W, C = vid.shape
    print(f"ğŸ“¼ ë¹„ë””ì˜¤ ì €ì¥ ì‹œì‘: {filename} (ì´ {T} í”„ë ˆì„, FPS: {frames_per_second})")

    process = (
        ffmpeg
        .input('pipe:', format='rawvideo', pix_fmt='rgb24', s=f'{W}x{H}', r=frames_per_second)
        .output(filename, pix_fmt='yuv420p', vcodec='libx264', r=frames_per_second)
        .overwrite_output()
        .run_async(pipe_stdin=True)
    )

    for frame in vid.numpy():
        process.stdin.write(frame.tobytes())
    process.stdin.close()
    process.wait()

# ğŸ”¹ ê°œë³„ ë¹„ë””ì˜¤ ì²˜ë¦¬ í•¨ìˆ˜
def preprocess_video(video_path, output_path, face_file):
    print(f"ğŸ¬ ë¹„ë””ì˜¤ ì²˜ë¦¬ ì‹œì‘: {video_path}")

    cap = cv2.VideoCapture(video_path)
    fps = int(round(cap.get(cv2.CAP_PROP_FPS)))
    frames = []

    # ğŸ”¥ Mediapipe ëœë“œë§ˆí¬ ê°ì§€ê¸° ì´ˆê¸°í™”
    landmarks_detector = LandmarksDetector()

    # ì–¼êµ´ ë°ì´í„° ë¡œë“œ
    face_data = load_face_data(face_file)

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            print("ğŸ“Œ ëª¨ë“  í”„ë ˆì„ ì²˜ë¦¬ ì™„ë£Œ!")
            break

        frame_number = int(cap.get(cv2.CAP_PROP_POS_FRAMES)) - 1
        print(f"ğŸ–¼ï¸ í”„ë ˆì„ {frame_number} ì²˜ë¦¬ ì¤‘...")

        # ê°€ì¥ ì‹ ë¢°ë„ ë†’ì€ í™”ì ì„ íƒ
        main_speaker = select_main_speaker(face_data, frame_number)

        if main_speaker:
            x1, y1, x2, y2 = map(int, main_speaker["bbox"])
            face_crop = frame[y1:y2, x1:x2]  # ì–¼êµ´ ì˜ì—­ í¬ë¡­
            print(f"ğŸ” í”„ë ˆì„ {frame_number}: ì–¼êµ´ í¬ë¡­ ì™„ë£Œ {x1, y1, x2, y2}")

            # ğŸ¯ ëœë“œë§ˆí¬ ê°ì§€ ìˆ˜í–‰
            landmarks_list = landmarks_detector.detect([frame], landmarks_detector.full_range_detector)

            if landmarks_list and landmarks_list[0] is not None:
                landmarks = landmarks_list[0]
                lip_region = extract_lip_region(frame, landmarks)

                if lip_region is not None:
                    frame = lip_region  # ì…ìˆ ë§Œ ì €ì¥
                    print(f"âœ… í”„ë ˆì„ {frame_number}: ì…ìˆ  í¬ë¡­ ì™„ë£Œ")

        frames.append(frame)
        frame_count += 1

    cap.release()

    if len(frames) == 0:
        print("ğŸš¨ ì €ì¥í•  í”„ë ˆì„ì´ ì—†ìŠµë‹ˆë‹¤. ë¹„ë””ì˜¤ ìƒì„± ì·¨ì†Œ.")
        return

    # ë¦¬ìŠ¤íŠ¸ -> NumPy ë°°ì—´ -> Torch Tensor ë³€í™˜
    frames_array = np.array(frames)
    torch_frames = torch.from_numpy(frames_array)

    save2vid(output_path, torch_frames, fps)
    print(f"âœ… ì…ìˆ  ì¶”ì¶œ ë¹„ë””ì˜¤ ì €ì¥ ì™„ë£Œ: {output_path}")

# ğŸ”¹ í´ë” ë‚´ ëª¨ë“  ë¹„ë””ì˜¤ ì²˜ë¦¬ í•¨ìˆ˜
def preprocess_video_folder(src_folder, dst_folder, face_file):
    os.makedirs(dst_folder, exist_ok=True)
    face_data = load_face_data(face_file)

    for filename in os.listdir(src_folder):
        if filename.lower().endswith(('.mp4', '.avi', '.mov', '.mkv')):
            src_path = os.path.join(src_folder, filename)
            dst_path = os.path.join(dst_folder, filename)
            print(f"ğŸ“‚ {src_path} -> {dst_path} ë³€í™˜ ì‹œì‘")
            preprocess_video(src_path, dst_path, face_data)

# âœ… **ì‹¤í–‰ ì˜ˆì‹œ**
video_path = "/content/Visual_Speech_Recognition_for_Multiple_Languages/Taylor Swift's inspirational speech.mp4"
output_path = "/content/processed_lipsxxxx.mp4"
face_file = "/content/Visual_Speech_Recognition_for_Multiple_Languages/faces (1).pckl"

preprocess_video(video_path, output_path, face_file)
```
